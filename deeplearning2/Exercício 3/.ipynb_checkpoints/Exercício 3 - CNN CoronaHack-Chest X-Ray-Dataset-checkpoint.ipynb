## Carga e Pr-Processamento das Imagens

Durante a anlise exploratria visual das imagens, percebemos que podemos aumentar contraste, brilho e sharpness nelas de modo  evidenciar detalhes e facilitar a fase de extrao de features na convoluo.

Como tambm j verificamos anteriormente, as classes no conjunto de treinamento esto desbalanceadas. 

Abaixo definimos algumas funes para carga e pr-processamento das imagens que sero utilizadas para o data augmentation e preparo das imagens para o modelo.

### Funo para Carga das Imagens

````python
# Carrega imagem para array e normaliza
def carrega_imagem(path, dimensao = DIM_IMAGEM):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, dimensao)
    
    img = np.expand_dims(img, axis=-1)
    
    # normaliza a imagem
    img = img / 255.0
    
    return img
````

### Funes de Pr-Processamento de Imagens

````python
# Funo para aumentar contraste da imagem

def aumentar_contraste(img, fator = 50):
    
    # Calcula a mdia e o desvio padro da imagem
    mean = np.mean(img)
    std = np.std(img)
    
    # Aplica uma transformao de contraste com base no fator fornecido
    img_contraste = (img - mean) / std * fator + mean
    
    # Limita os valores aos limites de 0 e 255
    img_contraste = np.clip(img_contraste, 0, 255)
    
    return img_contraste

# Funo para aumentar brilho da imagem

def aumentar_brilho(img, fator = 0.2):
    
    img_brilho = img + fator
    
    img_brilho = np.clip(img_brilho, 0, 255)
    
    return img_brilho

# Funo para melhorar o sharpness da imagem

def aumentar_sharpness(img, fator = 5):
    # ref: https://setosa.io/ev/image-kernels/
    
    kernel = np.array([[0, -1, 0],
                       [-1, fator, -1],
                       [0, -1, 0]])
    
    img = img.squeeze()
    img_sharp = convolve2d(img, kernel, mode = 'same', boundary = 'fill', fillvalue = 0)    
    img = img[:, :, np.newaxis]

    return img_sharp

# Funo para Flipar Horizontalmente uma imagem
def flip_horizontal(img):
    return np.fliplr(img)
````

## Criao do modelo a partir da LetNet-5

Iniciamos os estudos a partir do modelo LetNet-5 e fomos modificando a medida que testvamos e avalivamos o desempenho a partir das mtricas como Matriz de Confuso, Curva Roc e Matriz de Classificao.

A arquitetura da rede aqui apresentada  o resultado deste estudo.

````python
# Inicializao
model = Sequential()

model.add(Conv2D(8, (4, 4), activation = 'sigmoid', input_shape = (DIM_IMAGEM[0], DIM_IMAGEM[1], 1)))
model.add(MaxPooling2D((3, 3)))
model.add(Dropout(0.1))

model.add(Conv2D(8, (4, 4), activation = 'sigmoid'))
model.add(AveragePooling2D((2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(8, (4, 4), activation = 'sigmoid'))
model.add(AveragePooling2D((2, 2)))
model.add(Dropout(0.1))

model.add(Conv2D(8, (4, 4), activation = 'sigmoid'))
model.add(MaxPooling2D((3, 3)))
model.add(Dropout(0.1))

model.add(Flatten())

model.add(Dense(256, activation = 'sigmoid'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))

model.add(Dense(1, activation = 'sigmoid'))
````

Tanto a funo **sigmoid** como a funo **ReLU** so funes no lineares. A escolha da funo **sigmoid** nas camadas de confoluo e na primeira camada densa se deu para ajudar o modelo a aprender representaes mais sutis dos dados, uma vez que as imagens podem conter diferenas muito pouco perceptveis entre um pulmo saudvel e um com pneumonia. A flexibilidade da funo **sigmoid** foi fundamental. Durante os estudos, as tentativas de utilizar **ReLU** nessas camadas resultaram numa significativa piora do modelo, mesmo assim, ao colocar a **ReLU** nas duas camadas densas de 128 e 64 neurnios evitou o modelo de atingir o overfit, que aconteceu diversas vezes quando usando tambm **sigmoid** nessas camadas.

Tambm, a escolha da funo **sigmoid** na camada de sada foi por conta de se tratar de uma tarefa de classificao binria. A funo mapeia a sada para uma probabilidade, entre 0 e 1.
