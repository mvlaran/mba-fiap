```python
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model, load_model

import numpy as np

#!pip install xgboost
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.manifold import TSNE

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, StratifiedKFold

from keras.callbacks import ModelCheckpoint
from keras.utils import plot_model

import seaborn as sns
```

```python
RANDOM_STATE = 73
```

 Carrega os dados do dataset MINST

```python
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
set(y_test)
```

```python
X_train.shape
```

```python
labels = [ 'T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle boot'  ]
```

```python
classes = set(y_test)

plt.figure(0)
for c in classes:
  idx =  np.where(y_test == c)
  if c + 1 == 10:
    break
  plt.subplot(430 + 1 + c)
  # plot raw pixel data
  plt.imshow(X_test[idx[0][0]], cmap=plt.get_cmap('gray'))

plt.show()
```

```python
X_train.shape
```

```python
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = RANDOM_STATE)

X_train = X_train.reshape(len(X_train), np.prod(X_train.shape[1:]))
X_test = X_test.reshape(len(X_test), np.prod(X_test.shape[1:]))
X_val = X_val.reshape(len(X_val), np.prod(X_val.shape[1:]))

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

```python
input_img= Input(shape=(784,))
encoded = Dense(units=32, activation='relu')(input_img)
decoded = Dense(units=784, activation='sigmoid')(encoded)
```

```python
autoencoder=Model(input_img, decoded)
autoencoder.summary()
```

```python
encoder = Model(input_img, encoded)
encoder.summary()
```

```python
checkpoint = ModelCheckpoint('/best_model_ac.h5', monitor='val_loss', save_best_only=True, verbose=1)

autoencoder.compile(optimizer='adam', loss='bce', metrics=['accuracy'])
```

```python
autoencoder.fit(X_train, X_train,
                epochs = 100, batch_size = 64,
                shuffle = True, validation_data = (X_val, X_val),
                callbacks=[checkpoint])
```

```python
best_autoencoder = load_model('/best_model_ac.h5')
```

```python
encoded_imgs = encoder.predict(X_test)
predicted = best_autoencoder.predict(X_test)
```

```python
encoded_imgs.shape
```

```python
plt.figure(figsize=(40, 4))
for i in range(10):
    # display original
    ax = plt.subplot(3, 20, i + 1)
    plt.imshow(X_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # display encoded image
    ax = plt.subplot(3, 20, i + 1 + 20)
    plt.imshow(encoded_imgs[i].reshape(8,4))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    # display reconstruction
    ax = plt.subplot(3, 20, 2*20 +i+ 1)
    plt.imshow(predicted[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)


plt.show()
```

```python
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)
```

# Processo de Classificaao com Random Forest

```python
# modelRf = RandomForestClassifier()
# modelRf.fit(X_train_encoded, y_train)
# y_pred = modelRf.predict(X_test_encoded)

# modelXG = XGBClassifier()
# modelXG.fit(X_train_encoded, y_train)
# y_pred = modelXG.predict(X_test_encoded)
```

```python
param_grid = {
    'n_estimators': [500, 750],
    'max_depth': [5, 7],
    'learning_rate': [0.1, 0.25, 0.5],
}

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state = RANDOM_STATE)
grid_search = GridSearchCV(estimator=XGBClassifier(), param_grid=param_grid, scoring='f1_macro', cv=cv, n_jobs=-1, verbose = 5)
grid_search.fit(X_train_encoded, y_train)
best_params = grid_search.best_params_
best_score = grid_search.best_score_
y_pred = grid_search.best_estimator_.predict(X_test_encoded)
```

```python
print(f'Best Params: {best_params}\nBest Score: {best_score}')
```

```python
print(classification_report(y_test, y_pred))

print(accuracy_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

cm = confusion_matrix(y_test, y_pred, labels=grid_search.classes_, normalize = 'true')
sns.heatmap(cm, annot=True, cmap="Blues", xticklabels=labels, yticklabels=labels)
plt.xlabel("Previses")
plt.ylabel("Valores Reais")
plt.title("Matriz de Confuso")
plt.show()
plt.show()
```

# Usando Autoencoder Convolucional

```python
# Para garantir o ambiente, recarrego os dados e refao as variveis

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = RANDOM_STATE)

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_val.shape
```

```python
# Redimensionando as imagens para o formato (28, 28, 1)
X_train = np.reshape(X_train, (len(X_train), 28, 28, 1))
X_test = np.reshape(X_test, (len(X_test), 28, 28, 1))
X_val = np.reshape(X_val, (len(X_val), 28, 28, 1))
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_val.shape
```

```python
input_img = Input(shape=(28, 28, 1))

x = Conv2D(16, (4, 4), activation='sigmoid', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (4, 4), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (4, 4), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(32, (4, 4), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(16, (4, 4), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(8, (4, 4), activation='sigmoid', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(1, (3, 3), activation='sigmoid', padding='valid')(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='valid')(x)
```

```python
autoencoder_aec = Model(input_img, decoded)
autoencoder_aec.summary()
```

```python
plot_model(autoencoder_aec, to_file = 'FashioMNIST_AEC.png', show_shapes = True)
```

```python
# Compilando e treinando o autoencoder
autoencoder_aec.compile(optimizer='adam', loss='bce', metrics=['accuracy'])
checkpoint = ModelCheckpoint('/best_model_aec.h5', monitor='val_loss', save_best_only=True, verbose=1)

treinamento = autoencoder_aec.fit(X_train, X_train,
                              epochs = 30, batch_size = 64, 
                              shuffle = True, validation_data = (X_val, X_val),
                              callbacks = [checkpoint])
```

```python
best_autoencoder_conv = load_model('/best_model_aec.h5')
```

```python
# Obtendo a sada da ltima camada de max pooling do encoder
encoded_features = best_autoencoder_conv.predict(X_train)

# Realizando um reshape manual para o formato (n_samples, n_features)
encoded_features = encoded_features.reshape(encoded_features.shape[0], -1)
```

```python
# Para fins de comparao, usamos os mesmos parmetros do modelo anterior
modelXG_aec = XGBClassifier(learning_rate = 0.1, max_depth = 7, n_estimators = 750, n_jobs = -1, random_state = RANDOM_STATE)
modelXG_aec.fit(encoded_features, y_train)
```

```python
# Obtendo a sada da ltima camada de max pooling do encoder para o conjunto de teste
encoded_features_test = best_autoencoder_conv.predict(X_test)

# Realizando um reshape manual para o formato (n_samples, n_features) para o conjunto de teste
encoded_features_test = encoded_features_test.reshape(encoded_features_test.shape[0], -1)

y_pred = modelXG_aec.predict(encoded_features_test)
```

```python
print(classification_report(y_test, y_pred))

print(accuracy_score(y_test, y_pred))
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

cm = confusion_matrix(y_test, y_pred, labels=grid_search.classes_, normalize = 'true')
sns.heatmap(cm, annot=True, cmap="Blues", xticklabels=labels, yticklabels=labels)
plt.xlabel("Previses")
plt.ylabel("Valores Reais")
plt.title("Matriz de Confuso")
plt.show()
plt.show()
```

```python
tsne = TSNE(n_iter = 2000)
X_embedded_tsne = tsne.fit_transform(encoded_features_test)
```

```python
sns.set(rc={'figure.figsize':(20,10)})
palette = sns.color_palette("bright", 10)
sns.scatterplot(x = X_embedded_tsne[:,0], y = X_embedded_tsne[:,1], hue=y_test, legend='full', palette=palette)
```
